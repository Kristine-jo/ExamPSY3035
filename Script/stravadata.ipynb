# My first hypothesis: Time of day effects the average pace on all my runs

```python

# Ensure "Activity Date" is in datetime format
df["Activity Date"] = pd.to_datetime(df["Activity Date"], errors='coerce')

# Now, you can extract the hour
df["hour"] = df["Activity Date"].dt.hour

# Check the first few rows
print(df.head())
```

```python
df["Activity Time"] = pd.to_datetime(df_runs["Activity Date"]) # Extracting date and time from Activity Date
df["hour"] = df["Activity Date"].dt.hour  # Extracting time in hours from Activity Date
```

```python
def classify_time(hour): #Classifies whether time of run is in the morning, evening or afternoon
    if hour < 13:
        return "Morning" #before 13.00
    elif hour >= 18:
        return "Evening" # after 18.00
    else:
        return "Afternoon"  # Handle hours between 13 and 18

df["time_of_day"] = df["hour"].apply(classify_time) #converts hours in to time of day(morning, afternoon, evning)


```

## boxplot for all my runs: 
The boxplot show there is no significant change in the runningpace, of all of my runs, based on time of day. i have an average just above 5.5 pace in the morning, afternoon and evening.

```python
#boxplot
plt.figure(figsize=(8, 5))
plt.grid(False) #removes the bacground lines
sns.set_style("whitegrid")  # Adds a clean background
sns.boxplot(
    x=df["time_of_day"], 
    y=df_runs["average_pace"], 
    color="Seagreen",  # Uses a more visually appealing color palette
    width=0.5,  # Adjusts box width for better spacing
    linewidth=1.0  # Makes the box edges more visible
)
plt.xlabel("") #didn't want the lable for x axis
plt.ylabel("Average Pace", fontsize=12, fontname="Times New Roman")
plt.title("Average Pace by Time of Day", fontsize=14, fontname="Times New Roman", fontweight="bold")

plt.xticks(fontsize=11, fontname="Times New Roman")
plt.yticks(fontsize=11, fontname="Times New Roman")

sns.despine()  # Removes top and right borders for a cleaner look
plt.show()

#boxplot for evning lacks upper whisker
#Im using seaborn to get nice boxplots. However, seaborn follows Tukeys rule for outliers, and does not include the "whiskers" if the outlier is to large(above Q3)

```

# Thrid hypothesis: Which distance did i run most frequently in 2024


```python
# Define the bins for categorizing the distances
bins = [0, 3, 6, 11]  # These are the upper bounds of each range (1-3 km, 4-6 km, 7-11 km), using upper bounds ensures that the categories are properly seperated.
labels = ['1-3 km', '4-6 km', '7-11 km']  # names for the categories

# Categorized the runs based on the distance
df_runs['Distance_category'] = pd.cut(df_runs['Distance'], bins=bins, labels=labels, right=True) #remember, df-runs is called this because this variable only contains runs.

# Display the updated DataFrame with the new category column
print(df_runs[['Distance', 'Distance_category']])

```

```python
# Count the number of runs in each distance category
category_counts = df_runs['Distance_category'].value_counts()

# Print the counts to verify
print(category_counts)
```

# third hypothesis: Which month of the year did i run the most in 2024

```python
#convert date in to datetime format
df["Activity Date"] = pd.to_datetime(df["Activity Date"]) #extract only time from the variable
```

```python

# Ensure 'Activity Date' is in datetime format
df['Activity Date'] = pd.to_datetime(df['Activity Date'])

# Extract abbreviated month names
df['Month'] = df['Activity Date'].dt.strftime('%b')  # 'Jan', 'Feb', etc.

# Count runs per month
monthly_counts = df['Month'].value_counts()

# Define month order in abbreviations
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_counts = monthly_counts.reindex(month_order, fill_value=0)

```

## Scatterplot
The scatterplot shows that in all of 2024, i had the highest number of runs in may

```python
# Set figure size
plt.figure(figsize=(12, 6))

# Define a soft color palette
scatter_color = "crimson"  # Elegant blue
line_color = "gray" # Soft gray

# Create scatter plot
sns.scatterplot(
    x=monthly_counts.index, 
    y=monthly_counts.values, 
    s=150, 
    color=scatter_color, 
    edgecolor="crimson", 
    linewidth=1.5, 
    marker="o",  # Circle markers
    zorder=3
)

# Connect points with a smooth line
plt.plot(
    monthly_counts.index, 
    monthly_counts.values, 
    linestyle="-", 
    linewidth=1.5, 
    color=line_color, 
    alpha=0.6, 
    zorder=2
)

# Add exact numbers above points
for i, value in enumerate(monthly_counts.values):
    plt.text(i, value + 1.5, str(value), ha="center", fontsize=12, fontweight="bold", color="black", fontname="Times New Roman")

# Labels and title with Times New Roman font
plt.xlabel("Month", fontsize=14, fontweight="bold", color="black", fontname="Times New Roman")
plt.ylabel("Number of Runs", fontsize=14, fontweight="bold", color="black", fontname="Times New Roman")
plt.ylim(-1, 18) #adds some space on the y axis so that the datapoints don‚Äôt align with the x-axis
plt.title("Number of Runs per Month", fontsize=16, fontweight="bold", color="black", fontname="Times New Roman")


# Adjust x and y ticks
plt.xticks(rotation=0, fontsize=12, color="black", fontname="Times New Roman")
plt.yticks(fontsize=12, color="black", fontname="Times New Roman")

# Subtle grid (only horizontal lines)
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Remove unnecessary spines for a cleaner look
sns.despine(left=True, bottom=True)

# Show the plot
plt.show()
```
