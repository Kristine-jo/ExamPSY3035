```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (update the path to match your file location)
df = pd.read_csv(r"C:\Users\krist\PSY-3035\github\ExamPSY3035\Data\runlog\my_running.csv")

# Convert activity_date to datetime format
df['Activity Date'] = pd.to_datetime(df['Activity Date'])
df["hour"] = df['Activity Date'].dt.hour  # Extract hour


```

```python
print(df.columns)  # Check column names
```

# My first hypothesis: Do i run faster in the evning than in the morning

```python
def classify_time(hour):
    if hour < 13:
        return "Morning"
    elif hour >= 18:
        return "Evening"
    else:
        return "Afternoon"  # Handle hours between 13 and 17

df["time_of_day"] = df["hour"].apply(classify_time)


```

```python
#This plot contains all my runs, and results with them being quite similar
```

```python
# Now, plot the boxplot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['time_of_day'], y=df['average_pace'])
plt.xlabel('Time of Day')
plt.ylabel('Average Pace (min/km)')
plt.title('Average Pace by Time of Day')
plt.show()
```

```python
df_4_6k = df[(df["Distance"] >= 4) & (df["Distance"] <= 6)]  # Keep runs between 4 and 6 km
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df_4_6k["time_of_day"], y=df_4_6k["Average Grade Adjusted Pace"])
plt.xlabel("Time of Day")
plt.ylabel("Average Grade Adjusted Pace (min/km)")
plt.title("Running Pace for 4-6K: Morning vs. Evening")
plt.grid()
plt.show()
#this figure includes afternoon, and i do not want it included
```

```python
df_4_6k_filtered = df_4_6k[df_4_6k["time_of_day"] != "Afternoon"]
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(x=df_4_6k_filtered["time_of_day"], y=df_4_6k_filtered["Average Grade Adjusted Pace"])
plt.xlabel("Time of Day")
plt.ylabel("Average Grade Adjusted Pace (min/km)")
plt.title("Running Pace for 4-6K: Morning vs. Evening")
plt.grid()
plt.show()
```
