```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (update the path to match your file location)
df = pd.read_csv(r"C:\Users\krist\PSY-3035\github\ExamPSY3035\Data\runlog\my_running.csv")

# Convert activity_date to datetime format
df['Activity Date'] = pd.to_datetime(df['Activity Date'])
df["hour"] = df['Activity Date'].dt.hour  # Extract hour


```

# My first hypothesis: Do i run faster in the evning than in the morning

```python
def classify_time(hour):
    if hour < 13:
        return "Morning"
    elif hour >= 18:
        return "Evening"
    else:
        return "Afternoon"  # Handle hours between 13 and 17

df["time_of_day"] = df["hour"].apply(classify_time)


```

```python
#This plot contains all my runs, and results with them being quite similar
```

```python
#because i have a substantial outlier, it makes the plot difficult to read. I therefore want to remove this
# Remove the row with index 26
df_runs = df_runs.drop(26)
```

```python
# Now, plot the boxplot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 5))
sns.boxplot(x=df_runs['time_of_day'], y=df_runs['average_pace'])
plt.xlabel('Time of Day')
plt.ylabel('Average Pace (min/km)')
plt.title('Average Pace by Time of Day')
plt.show()
```

# Thrid hypothesis:which distance do i run most frequently?

```python
# Define the bins for categorizing the distances
bins = [0, 3, 6, 11]  # These are the upper bounds of each range (1-3 km, 4-6 km, 7-11 km)
labels = ['1-3 km', '4-6 km', '7-11 km']  # Labels for the categories

# Categorize the runs based on the distance
df_runs['Distance_category'] = pd.cut(df_runs['Distance'], bins=bins, labels=labels, right=True)

# Display the updated DataFrame with the new category column
print(df_runs[['Distance', 'Distance_category']])

```

```python
# Count the number of runs in each distance category
category_counts = df_runs['Distance_category'].value_counts()

# Print the counts to verify
print(category_counts)
```

```python
# Plotting a pie chart
plt.figure(figsize=(7, 7))
category_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('Blues_d', n_colors=len(category_counts)))

# Adding a title
plt.title("Run Distribution by Distance Category")

# Show the plot
plt.show()
```

```python
#Because the piechart is not optimal on infromation, meaning we cant actually see the number of runs, im making an additional chart
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(8, 6))

# Create barplot with hue explicitly set
sns.barplot(x=category_counts.index, y=category_counts.values, hue=category_counts.index, palette='Blues_d', legend=False)

# Adding labels and title
plt.xlabel("Distance Category")
plt.ylabel("Number of Runs")
plt.title("Frequency of Runs by Distance Category")

# Show the plot
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure 'Activity Date' is in datetime format
df['Activity Date'] = pd.to_datetime(df['Activity Date'])

# Extract abbreviated month names
df['Month'] = df['Activity Date'].dt.strftime('%b')  # 'Jan', 'Feb', etc.

# Count runs per month
monthly_counts = df['Month'].value_counts()

# Define month order in abbreviations
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_counts = monthly_counts.reindex(month_order, fill_value=0)

# Create bar plot
plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_counts.index, y=monthly_counts.values, hue=monthly_counts.index, legend=False, palette="Blues_d")
plt.xlabel("Month")
plt.ylabel("Number of Runs")
plt.title("Number of Runs per Month")
plt.show()
```
