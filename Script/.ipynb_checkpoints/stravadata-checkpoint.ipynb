### Converting and calculating variables
Before i could start exploring my hypothesis, i needed to both extract and convert some variables form the dataset. 
First, Stava only keeps the record of average graded pace, and not average pace, however i wanted to investigate the average pace. To do this i first had to convert the time variable from seconds to minutes. Then i had to calculate the average pace using minutes and ditance in km, which was already in the datafile.

```python
#want to convert time from seconds to minutes, and add avrage pace
df["time_minutes"] = df["Elapsed Time"] / 60 #convert fromm seconds to minutes
df["average_pace"] = df["time_minutes"] / df["Distance"] #adds average pace
print(df)
```

# My first hypothesis: Time of day effects the average pace on all my runs

In my first hypothesis i wanted to investigate whethere there is a time og day where i acheive better results while running, and whether there is a time of day where i achieve worse results.

```python

# Ensure "Activity Date" is in datetime format
df["Activity Date"] = pd.to_datetime(df["Activity Date"], errors='coerce') #coerce make sure invalid numers are not included

# extracting time of day
df["hour"] = df["Activity Date"].dt.hour

# Check the first few rows
print(df.head())
```

```python
#df["Activity Time"] = pd.to_datetime(df_runs["Activity Date"]) # Extracting date and time from Activity Date
#df["hour"] = df["Activity Date"].dt.hour  # Extracting time in hours from Activity Date
```

```python
#Classifies whether time of run is in the morning, evening or afternoon
def classify_time(hour): 
    if hour < 13:
        return "Morning" # if run is before before 13.00
    elif hour >= 18:
        return "Evening" #  if run is after 18.00
    else:
        return "Afternoon"  # Handle hours between 13.00 and 18.00

df["time_of_day"] = df["hour"].apply(classify_time) #converts hours in to time of day(morning, afternoon, evening)


```

```python
print(df["time_of_day"]) #Showing the convertion
```

```python
# Because I had a substantial outlier, it makes the plot difficult to read. I therefore want to remove this. (It was a mountain run, wich lies in the run category)
df_runs = df_runs.drop(26)  # Remove the row with index 26
```

## boxplot for all my runs: 
The boxplot show there is no significant change in the runningpace, of all of my runs, based on time of day. I have an average just above 5.5 pace in the morning, afternoon and evening. So based on the plot i can say, time of day does not effect the results of my run.

The boxplot figures are nice for comparing averages sucha as these. It gives a clear picture of what the average pace is, but also the distibution of data within each category. 

To have a clean, elegant look, which allows for simple eterpretation, i chose a calm color, and a "clean" easy to read font, with thin lines tracing the figures. In order to make the plot easily interpreteble i added some gray lines in the back. So you can quickly see exactly which pace on the y-axsis the average pace in the boxplot aligns with. 



```python
#boxplot for all runs
plt.figure(figsize=(5, 5))
plt.grid(axis='y', linestyle="--", alpha=0.7, color="gray") #adds grey lines to the plot
sns.set_style("whitegrid")  # Adds a clean background
sns.boxplot(
    x=df["time_of_day"], 
    y=df_runs["average_pace"], 
    color="cornflowerblue",  # Uses a more visually appealing color palette
    width=0.5,  # Adjusts box width for better spacing
    linewidth=1.0  # Makes the box edges more visible
)
plt.xlabel("") #didn't want the lable for x axis
plt.ylabel("Average Pace", fontsize=12, fontname="Times New Roman")
plt.title("Average Pace by Time of Day", fontsize=14, fontname="Times New Roman", fontweight="bold")

plt.xticks(fontsize=11, fontname="Times New Roman")
plt.yticks(fontsize=11, fontname="Times New Roman")

sns.despine()  # Removes top and right borders for a cleaner look
plt.show()

#boxplot for evning lacks upper whisker
#Im using seaborn to get nice boxplots. However, seaborn follows Tukeys rule for outliers, and does not include the "whiskers" if the outlier is to large(above Q3)

```

# Second hypothesis: Time of day effects the pace on my 4-6km runs

After finding no substential diffrence between all my runs depending on time of day, i wanted to further investigate if a effect could be found between time of day and my 4-6km runs. 

```python
df_4_6k = df[(df["Distance"] >= 4) & (df["Distance"] <= 6)]  # Keep runs between 4 and 6 km
```

## Boxplot for 4-6km runs:
The 4-6km runs show a bigger mean diffrence between time of day and pace of run. In the morning I run the fastest, with a mean pace just below 5.5 minutes per kilometer. In the evening the pace is a bit slower, with just about ca 5.6 minutes per kilometer. I run the slowest in the afternoon when running 4-6km, here the pase is closer to 6. 

To make this plot easily compareble with the boxplot containing all my runs, i want them to be alike, exept for in color. Because i want them to be compareble i will use the same values on the y-axsis.


```python

# Set Seaborn style
sns.set_style("whitegrid")

# Create figure
plt.figure(figsize=(5, 5))

#boxplot
sns.boxplot(
    x=df_4_6k["time_of_day"], 
    y=df_4_6k["average_pace"], 
    color="#F4A460", 
    width=0.5,
    linewidth=1.0,  
    flierprops={'marker': 'o', 'markersize': 6, 'markerfacecolor': 'black', 'markeredgecolor': 'black'}
)

# Labels and title
plt.xlabel("")  # Remove "Time of Day" label
plt.ylabel("Average Pace (min/km)", fontsize=14, fontname="Times new roman")
plt.title("Running Pace for 4-6K: Morning vs. Evening", fontsize=14, fontname="Times new roman", fontweight="bold")

plt.xticks(fontname="Times new roman", fontsize=14)
plt.yticks(fontname="Times new roman", fontsize=14)
plt.ylim(4.5, 7.5) #using same values on the y-axis as in the boxplots above

# Improve grid visibility
plt.grid(axis='y', linestyle="--", alpha=0.7, color="gray")

# Remove top and right spines (borders)
sns.despine(top=True, right=True)

# Show plot
plt.show()
```

# Thrid hypothesis: Which distance did i run most frequently in 2024
Now i want to find out which of my runningdistances i ran the most frequenlty in 2024. I will categorize the runs in to bins of 1-3km, 4-6km and 7-11km. However, i should note that in the colder months (january, february,march, april) i ran mostly on on the treadmil, where mostly havent recorded my runs. So these numbers, represent largely my "outside running".

```python
# Define the bins for categorizing the distances
bins = [0, 3, 6, 11]  # These are the upper bounds of each range (1-3 km, 4-6 km, 7-11 km), using upper bounds ensures that the categories are properly seperated.
labels = ['1-3 km', '4-6 km', '7-11 km']  # names for the categories

# Categorized the runs based on the distance
df_runs['Distance_category'] = pd.cut(df_runs['Distance'], bins=bins, labels=labels, right=True) #remember, df-runs is called this because this variable only contains runs.

# Display the updated DataFrame with the new category column
print(df_runs[['Distance', 'Distance_category']])

```

```python
# Count the number of runs in each distance category
category_counts = df_runs['Distance_category'].value_counts()

# Print the counts to verify
print(category_counts)
```

## Pie chart
The pie chart show clearly that there are far more 4-6k runs. The chart show that 60% of all my runs in 2024 are 4-6km. Next is the 7-11 km runs, which cover 32% of the chart. The distnce that i have run the least is 1-3km runs covering just 8% of the 2024 runs. 
Using the piechart gives a visually intuetiv understanding of how big a portion my runs are compared to one another. I added a colormap in a legend to make it easy to understand which category is which, without having to much text surrounding the piechart itself. For esthetic reasons i have uses a white, to display the prosentages, i think this is the more estetchically pleasing color, but to make sure the prosentages are clearly wisible i have made their fontsize quite big.



```python

# Example data (replace with actual data)
category_counts = pd.Series([4, 30, 16], index=["1-3 km", "4-6 km", "7-11 km"])

# Define color palette
colors = sns.color_palette("pastel", n_colors=len(category_counts)) 

# Create figure
plt.figure(figsize=(7, 7))

# Plot pie chart without category labels
wedges, _, autotexts = plt.pie(
    category_counts, 
    autopct='%1.1f%%', 
    colors=colors, 
    textprops={"fontname": "Times new roman", "fontweight": "bold", "fontsize":"20", "color":"white" },
    wedgeprops={"edgecolor": "white"}  # White line between the categories
)

#colormap outside the pie chart
plt.legend(wedges, category_counts.index, title="Distance Categories",
           loc="center left", bbox_to_anchor=(1, 0.5),
           prop={"family": "Times new roman", "size": 12},
          title_fontproperties={"family": "Times new roman", "size": 12}) #this changes the title font in the ledgend

# Adding a title
plt.title("Distribution of my running in 2024", fontname="Times new roman", fontweight="bold", fontsize=16)

# Show the plot
plt.show()
```

## Barplot
The pie chart show us the prosentage, however we also want to see the spesific number of runs, and i will do this trough a barplot.
The barchart show that out of all my 50 runs in 2024, 30 of them where 4-6km distances, 16 where 7-11km and 4 where 1-3km.  

The barplot contains a legend giving the total number of runs, making it easier to understand just how much 30 runs are compared to total number of run. While each bar have their belonging category of distance. I have the same color for each bar to match the piechart. 

```python
# Set figure size
plt.figure(figsize=(6, 5))

# Barplot
sns.barplot(
    x=category_counts.index, 
    y=category_counts.values, 
    hue=category_counts.index, 
    legend=False,
    palette=sns.color_palette("pastel", n_colors=len(category_counts))
)

# Calculate total runs
total_runs = category_counts.sum()

# Add the total runs as a label in the legend
handles, labels = plt.gca().get_legend_handles_labels()


# Add a custom legend for "Total Runs"
handles.append(plt.Line2D([0], [0], marker='o', color='w', label=f"Total Runs: {total_runs}", 
                          markerfacecolor='gray', markersize=10))
labels.append(f"Total Runs: {total_runs}")

# Customize the legend to only display total runs
plt.legend(handles=handles, labels=labels, loc="upper right", title="", 
           fontsize=12, title_fontsize=14, prop={'family': 'Times New Roman'})

# Adding labels and title
plt.xlabel("Distance Category", fontname="Times New Roman")
plt.ylabel("Number of Runs", fontname="Times New Roman")
plt.title("Frequency of Runs", fontname="Times New Roman", fontsize=16)

# Customizing tick labels and grid
plt.xticks(fontname="Times New Roman")
plt.yticks(fontname="Times New Roman")
plt.grid(axis='y', linestyle="--", alpha=0.7, color="gray")
plt.ylim(0, 50)
sns.despine(top=True, right=True)

# Show the plot
plt.show()

```

# third hypothesis: Which month of the year did i run the most in 2024

```python
#convert date in to datetime format
df["Activity Date"] = pd.to_datetime(df["Activity Date"]) #extract only time from the variable
```

```python

# Ensure 'Activity Date' is in datetime format
df['Activity Date'] = pd.to_datetime(df['Activity Date'])

# Extract abbreviated month names
df['Month'] = df['Activity Date'].dt.strftime('%b')  # 'Jan', 'Feb', etc.

# Count runs per month
monthly_counts = df['Month'].value_counts()

# Define month order in abbreviations
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_counts = monthly_counts.reindex(month_order, fill_value=0)

```

## Scatterplot
The scatterplot shows that in all of 2024, i had the highest number of runs in may. I chose a barplot because this neatly manages to present al the months and the belonging number of runs, without it being messy. Its clean because it creates a simple shape with the lines to each dot, making it intuetive to understand. 

```python
# Set figure size
plt.figure(figsize=(12, 6))

# Define a soft color palette
scatter_color = "cornflowerblue"  
line_color = "gray"

# Create scatter plot
sns.scatterplot(
    x=monthly_counts.index, 
    y=monthly_counts.values, 
    s=150, 
    color=scatter_color, 
    edgecolor="cornflowerblue", 
    linewidth=1.5, 
    marker="o",  # Circle markers
    zorder=3
)

# Connect points with a smooth line
plt.plot(
    monthly_counts.index, 
    monthly_counts.values, 
    linestyle="-", 
    linewidth=1.5, 
    color=line_color, 
    alpha=0.6, 
    zorder=2
)

# Add exact numbers above points (unceirtain if i should include)
#for i, value in enumerate(monthly_counts.values):
    #plt.text(i, value + 1.5, str(value), ha="center", fontsize=12, fontweight="bold", color="black", fontname="Times New Roman")

# Labels and title with Times New Roman font
plt.xlabel("Month", fontsize=14, fontweight="bold", color="black", fontname="Times New Roman")
plt.ylabel("Number of Runs", fontsize=14, fontweight="bold", color="black", fontname="Times New Roman")
plt.ylim(-1, 15) #adds some space on the y axis so that the datapoints don‚Äôt align with the x-axis
plt.title("Number of Runs per Month", fontsize=16, fontweight="bold", color="black", fontname="Times New Roman")

# Adjust x and y ticks
plt.xticks(rotation=0, fontsize=12, color="black", fontname="Times New Roman")
plt.yticks(fontsize=12, color="black", fontname="Times New Roman")

#background lines
plt.grid(axis="y", linestyle="--", alpha=0.5)

sns.despine(left=True, bottom=True)

# Show the plot
plt.show()
```
